### Let's read the data stored previously and start building the model

```python
import pandas as pd

weather_train = pd.read_csv('weather_train.csv')
X_train = weather_train.drop('RainTomorrow', axis=1)
y_train = weather_train['RainTomorrow']
X_train.shape, y_train.shape
```

```python
import pandas as pd

weather_test = pd.read_csv('weather_test.csv')
X_test = weather_test.drop('RainTomorrow', axis=1)
y_test = weather_test['RainTomorrow']
X_test.shape, y_test.shape
```

### Checking with Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=10)
dt.fit(X_train, y_train)
dt.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report

y_pred = dt.predict(X_test)

print(classification_report(y_test, y_pred))
```

### GridSearch CV for hypertune

```python
from sklearn.model_selection import GridSearchCV

grid_param = {
                'criterion': ['gini', 'entropy'],
                'max_depth' : range(2,20,2),
                'min_samples_leaf' : range(1,10,2),
                'min_samples_split': range(2,10,2),
                'splitter' : ['best', 'random']
    
}
grid_param
```

```python
grid_search = GridSearchCV(estimator=dt, param_grid=grid_param, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
dt_optimized = grid_search.best_estimator_
dt_optimized
```

```python
dt_optimized.fit(X_train, y_train)
y_pred = dt_optimized.predict(X_test)

print(classification_report(y_test, y_pred))
```

> Accuracy has improved from 0.74 to 0.83 but the recall of class 1 has fallen from 0.44 to 0.40

## Lets see how a random forest model would perform

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=10)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
from datetime import datetime

def timer(start_time=None):
    if not start_time:
        start_time = datetime.now()
        print('start time:', start_time)
        return start_time
    elif start_time:
        thour, temp_sec = divmod((datetime.now() - start_time).total_seconds(), 3600)
        tmin, tsec = divmod(temp_sec, 60)
        print('\n Time taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))
```

```python
## Hyper Parameter Optimization

params={
 "learning_rate"    : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
 "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
 "min_child_weight" : [ 1, 3, 5, 7 ],
 "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
 "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ]
    
}
```

```python
xgb = xgboost.XGBClassifier(random_state=10)
random_search = RandomizedSearchCV(estimator=xgb, param_distributions=params, n_iter=100, scoring='roc_auc',
                                      n_jobs=-1, cv=5, verbose=1)

start_time = timer()
random_search.fit(X_train, y_train)
timer(start_time)
```

```python
xgb_optimized = random_search.best_estimator_
xgb_optimized
```

```python
xgb_optimized.fit(X_train, y_train)
y_pred = xgb_optimized.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
tree.plot_tree(dt,filled=True)
```

```python
path = dt.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
ccp_alphas
```

```python
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)
print("Number of nodes in the last tree is: {} with ccp_alpha: {}".format(
      clfs[-1].tree_.node_count, ccp_alphas[-1]))
```

```python
train_scores = [clf.score(X_train, y_train) for clf in clfs]
test_scores = [clf.score(X_test, y_test) for clf in clfs]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas[:-1], train_scores[:-1], marker='o', label="train",
        drawstyle="steps-post")
ax.plot(ccp_alphas[:-1], test_scores[:-1], marker='o', label="test",
        drawstyle="steps-post")
ax.legend()
plt.show()
```
